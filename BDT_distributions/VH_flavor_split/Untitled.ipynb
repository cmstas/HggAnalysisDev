### MVA Scores after split

```python
df_with_mva = pd.read_pickle("../../MVAs/output/HHggTauTau_InclusivePresel_genZStudy_20210225_with_scores.pkl")
```

```python
ee_MVA_score = Hist1D(df_with_mva.loc[df_with_mva["genZ_decayMode"] == 1, "mva_score"].values, bins = np.linspace(0,1,20), label = "ee_events")
mm_MVA_score = Hist1D(df_with_mva.loc[df_with_mva["genZ_decayMode"] == 2, "mva_score"].values, bins = np.linspace(0,1,20), label = "mm_events")
tt_MVA_score = Hist1D(df_with_mva.loc[df_with_mva["genZ_decayMode"] == 3, "mva_score"].values, bins = np.linspace(0,1,20), label = "tautau_events")
others_MVA_score = Hist1D(df_with_mva.loc[df_with_mva["genZ_decayMode"] == 0, "mva_score"].values, bins = np.linspace(0,1,20), label = "ZToHad_events")
```
